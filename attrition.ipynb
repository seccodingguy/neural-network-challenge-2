## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
print(attrition_df.dtypes)

# Determine the number of unique values in each column
attrition_df.nunique().sort_values(ascending=False)
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
selected_columns = [
    "Age","DistanceFromHome","Education","EnvironmentSatisfaction",
    "HourlyRate","JobInvolvement","JobLevel","JobSatisfaction",
    "NumCompaniesWorked","PercentSalaryHike"
]

# Create X_df using your selected columns
X_df = attrition_df[selected_columns]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2, random_state=37)
```

```python
# Convert your X data to numeric data types however you see fit
X_numeric = pd.get_dummies(X_df)

# Add new code cells as necessary
X_numeric.head()
```

```python
# Create a StandardScaler
std_scaler = StandardScaler()

# Fit the StandardScaler to the training data
std_scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = std_scaler.transform(X_train)
X_test_scaled = std_scaler.transform(X_test)

```

```python
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder for the Department column
dept_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
one_hot_dept_encoded = dept_encoder.fit(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_dept_encoded = dept_encoder.transform(y_train[['Department']])
y_test_dept_encoded = dept_encoder.transform(y_test[['Department']])

display(y_train_dept_encoded)
display(y_test_dept_encoded)
```

```python
# Create a OneHotEncoder for the Attrition column
attr_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
attr_encoder.fit(y_train[['Attrition']])


# Create two new variables by applying the encoder
# to the training and testing data
y_train_attr_encoded = attr_encoder.transform(y_train[['Attrition']])
y_test_attr_encoded = attr_encoder.transform(y_test[['Attrition']])

display(y_train_attr_encoded)
display(y_test_attr_encoded)
```

## Part 2: Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
x_input_dim = X_train_scaled.shape[1]

# Create the input layer
input_layer = layers.Input(shape=(x_input_dim,))

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(32, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
dept_hidden_layer = layers.Dense(16, activation='relu')(shared_layer2)

# Create the output layer
dept_output_layer = layers.Dense(y_train_dept_encoded.shape[1], activation='softmax', name='department_output')(dept_hidden_layer)

```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attr_hidden_layer = layers.Dense(16, activation='relu')(shared_layer2)

# Create the output layer
attr_output_layer = layers.Dense(y_train_attr_encoded.shape[1], activation='softmax', name='attrition_output')(attr_hidden_layer)
```

```python
# Create the model
new_model = Model(inputs=input_layer, outputs=[dept_output_layer, attr_output_layer])

# Compile the model
new_model.compile(optimizer='adam', 
              loss={'department_output': 'categorical_crossentropy', 'attrition_output': 'categorical_crossentropy'}, 
              metrics={'department_output': 'accuracy', 'attrition_output': 'accuracy'})

# Summarize the model
new_model.summary()
```

```python
# Train the model
training = new_model.fit(
    X_train_scaled, 
    {'department_output': y_train_dept_encoded, 'attrition_output': y_train_attr_encoded},
    epochs=50,
    batch_size=32,
    validation_data=(X_test_scaled, {'department_output': y_test_dept_encoded, 'attrition_output': y_test_attr_encoded}),
    verbose=1
)
```

```python
# Evaluate the model with the testing data
new_model_evaluation = new_model.evaluate(
    X_test_scaled,
    {'department_output': y_test_dept_encoded, 'attrition_output': y_test_attr_encoded},
    verbose=1
)
```

```python
# Print the accuracy for both department and attrition
print(new_model_evaluation)

# Extract the specific metrics
test_loss = new_model_evaluation[0]
department_output_loss = new_model_evaluation[1]
attrition_output_loss = new_model_evaluation[2]

print(f"Test Loss: {test_loss}")
print(f"Test Department Output Loss: {department_output_loss}")
print(f"Test Attrition Output Loss: {attrition_output_loss}")

# Calculate accuracy for the department output
dept_pred = new_model.predict(X_test_scaled)[0]
dept_accuracy = np.mean(np.argmax(dept_pred, axis=1) == np.argmax(y_test_dept_encoded, axis=1))

# Calculate accuracy for the attrition output
attr_pred = new_model.predict(X_test_scaled)[1]
attr_accuracy = np.mean(np.argmax(attr_pred, axis=1) == np.argmax(y_test_attr_encoded, axis=1))

print(f"Test Department Output Accuracy: {dept_accuracy}")
print(f"Test Attrition Output Accuracy: {attr_accuracy}")
```
